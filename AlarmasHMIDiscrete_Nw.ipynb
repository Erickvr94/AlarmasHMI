```python
# ------------------------------------------------------------------------------
# BITS DE ENTRADA POR CIRCUITO (orden fijo del hardware)
# ------------------------------------------------------------------------------
CIRCUIT_BITS = [29, 30, 31, 16, 17, 18, 19, 20]

# ------------------------------------------------------------------------------
# FUNCIÓN: construir una fila con los valores por defecto del formato HMI
# ------------------------------------------------------------------------------
def make_row(name, alarm_text, clase, trigger_tag, trigger_bit):
    return {
        'ID': 0,
        'Name': name,
        'Alarm text [en-US], Alarm text 1': alarm_text,
        'FieldInfo [Alarm text 1]': np.nan,
        'Class': clase,
        'Trigger tag': trigger_tag,
        'Trigger bit': int(trigger_bit),
        'Trigger mode': 'On rising edge',
        'Acknowledgement tag': '<No value>',
        'Acknowledgement bit': 0,
        'PLC acknowledgement tag': '<No value>',
        'PLC acknowledgement bit': 0,
        'Priority': 0,
        'Info text [en-US], Info text': '<No value>',
        'Additional text 1 [en-US], Alarm text 2': '<No value>',
        'FieldInfo [Alarm text 2]': np.nan,
        'Additional text 2 [en-US], Alarm text 3': '<No value>',
        'FieldInfo [Alarm text 3]': np.nan,
        'Additional text 3 [en-US], Alarm text 4': '<No value>',
        'FieldInfo [Alarm text 4]': np.nan,
        'Additional text 4 [en-US], Alarm text 5': '<No value>',
        'FieldInfo [Alarm text 5]': np.nan,
        'Additional text 5 [en-US], Alarm text 6': '<No value>',
        'FieldInfo [Alarm text 6]': np.nan,
        'Additional text 6 [en-US], Alarm text 7': '<No value>',
        'FieldInfo [Alarm text 7]': np.nan,
        'Additional text 7 [en-US], Alarm text 8': '<No value>',
        'FieldInfo [Alarm text 8]': np.nan,
        'Additional text 8 [en-US], Alarm text 9': '<No value>',
        'FieldInfo [Alarm text 9]': np.nan,
        'Additional text 9 [en-US], Alarm text 10': '<No value>',
        'FieldInfo [Alarm text 10]': np.nan,
        'Alarm parameter 1': '<No value>',
        'Alarm parameter 2': '<No value>',
        'Alarm parameter 3': '<No value>',
        'Alarm parameter 4': '<No value>',
        'Alarm parameter 5': '<No value>',
        'Alarm parameter 6': '<No value>',
        'Alarm parameter 7': '<No value>',
        'Alarm parameter 8': '<No value>',
        'Alarm parameter 9': '<No value>',
        'Alarm parameter 10': '<No value>',
        'Area': 'HMI_RT_1::Alarming',
        'Origin': np.nan,
        'Audit class': '<No value>',
    }

# ------------------------------------------------------------------------------
# FUNCIÓN: generar todas las filas para un tablero dado
# ------------------------------------------------------------------------------
def generar_filas_tablero(pref, ta, is_tcp, cant_circ):
    """
    pref     : prefijo, p.ej. "CRI-AGROB"
    ta       : nombre del tablero, p.ej. "PS27_TA001"
    is_tcp   : bool
    cant_circ: int, número de circuitos
    """
    filas = []
    p = pref

    # — Circuitos On/Off ——————————————————————————————————————————————————————
    for i in range(1, cant_circ + 1):
        bit = CIRCUIT_BITS[i - 1]
        filas.append(make_row(
            f'{p}_Circuito_On_Off_{ta}_{i}',
            f'Circuito {i} {ta}_{i}',
            f'{p}_Circ_On_Off',
            f'{ta}.Inputs', bit
        ))

    # — Desactiva GM ——————————————————————————————————————————————————————————
    for i in range(1, cant_circ + 1):
        filas.append(make_row(
            f'{p}_Desactiva_GM_{ta}_{i}',
            f'Desactiva Circuito {i} {ta}_{i}',
            f'{p}_Operacion',
            f'{ta}.DesactivaGM', i - 1
        ))

    # — Falla Guardamotor —————————————————————————————————————————————————————
    filas.append(make_row(
        f'{p}_Falla_Guardamotor_{ta}',
        f'Falla Guardamotor {ta}',
        f'{p}_Alarm_Guarda',
        f'{ta}.Inputs', 28
    ))

    # — Supervisores ——————————————————————————————————————————————————————————
    filas.append(make_row(
        f'{p}_Supervisor_Volt_{ta}',
        f'Supervisor Voltaje {ta}',
        f'{p}_Alarm_Sup_Volt',
        f'{ta}.Inputs', 27
    ))
    filas.append(make_row(
        f'{p}_Supervisor_Trans_{ta}',
        f'Supervisor Transientes {ta}',
        f'{p}_Alarm_Sup_Tras',
        f'{ta}.Inputs', 26
    ))

    # — Paro Emergencia ———————————————————————————————————————————————————————
    filas.append(make_row(
        f'{p}_Paro_Emergencia_{ta}',
        f'Paro Emergencia {ta}',
        f'{p}_Stop_Parada',
        f'{ta}.Inputs', 11
    ))

    # — Comunicaciones ————————————————————————————————————————————————————————
    filas.append(make_row(
        f'{p}_Stat_Comm_{ta}',
        f'Stat Comm {ta}',
        f'{p}_Comunicaciones',
        f'{ta}.Stat_Comm', 0
    ))
    filas.append(make_row(
        f'{p}_Comm_Ant_{ta}',
        f'Comm Ant {ta}',
        f'{p}_Umbral_Antenas',
        f'Ant_Stat_{ta}', 0
    ))

    # — Variables exclusivas de tableros TCP ——————————————————————————————————
    if is_tcp:
        filas.append(make_row(
            f'{p}_Stat_Comm_MB_{ta}',
            f'Stat Comm MB {ta}',
            f'{p}_Comunicaciones',
            f'{ta}.Stat_Comm_MB', 0
        ))

        # Pasos Config TCP (12 pasos)
        for i in range(1, 13):
            filas.append(make_row(
                f'{p}_Pasos_Config_TCP_{ta}_{i}',
                f'Pasos Config TCP {ta} Paso {i}',
                f'{p}_Operacion',
                f'{ta}.Pasos_Habilit', i - 1
            ))

        # Pasos Activos TCP (12 pasos)
        for i in range(1, 13):
            filas.append(make_row(
                f'{p}_Pasos_Activos_TCP_{ta}_{i}',
                f'Pasos Activos TCP {ta} Paso {i}',
                f'{p}_Operacion',
                f'{ta}.Pasos_Acti', i - 1
            ))

        # Alarmas parámetros eléctricos — Alarm1
        for name, text, bit in [
            (f'{p}_Alarma_Temp_TCP_{ta}',           f'Alarma Temperatura TCP {ta}',                      0),
            (f'{p}_Alarma_Armonicos_V_L1_TCP_{ta}', f'Alarma Armonicos Voltaje Fase L1 TCP {ta}',       10),
            (f'{p}_Alarma_Armonicos_V_L2_TCP_{ta}', f'Alarma Armonicos Voltaje Fase L2 TCP {ta}',       11),
            (f'{p}_Alarma_Armonicos_V_L3_TCP_{ta}', f'Alarma Armonicos Voltaje Fase L3 TCP {ta}',       12),
            (f'{p}_Alarma_Voltaje_L1L2_TCP_{ta}',   f'Alarma Voltaje L1L2 TCP {ta}',                    13),
            (f'{p}_Alarma_Voltaje_L2L3_TCP_{ta}',   f'Alarma Voltaje L2L3 TCP {ta}',                    14),
            (f'{p}_Alarma_Voltaje_L3L1_TCP_{ta}',   f'Alarma Voltaje L3L1 TCP {ta}',                    15),
            (f'{p}_Alarma_Cos_L1_TCP_{ta}',         f'Alarma Cos Fase L1 TCP {ta}',                     26),
            (f'{p}_Alarma_Cos_L2_TCP_{ta}',         f'Alarma Cos Fase L2 TCP {ta}',                     27),
            (f'{p}_Alarma_Cos_L3_TCP_{ta}',         f'Alarma Cos Fase L3 TCP {ta}',                     28),
        ]:
            filas.append(make_row(name, text, f'{p}_Alarmas', f'{ta}.Alarm1', bit))

        # Alarmas parámetros eléctricos — Alarm2
        for name, text, bit in [
            (f'{p}_Alarma_Armonicos_I_L1_TCP_{ta}', f'Alarma Armonicos Corriente Fase L1 TCP {ta}',     3),
            (f'{p}_Alarma_Armonicos_I_L2_TCP_{ta}', f'Alarma Armonicos Corriente Fase L2 TCP {ta}',     4),
            (f'{p}_Alarma_Armonicos_I_L3_TCP_{ta}', f'Alarma Armonicos Corriente Fase L3 TCP {ta}',     5),
            (f'{p}_Alarma_Energia_Inductiva_TCP_{ta}',  f'Alarma Energia Inductiva TCP {ta}',           10),
            (f'{p}_Alarma_Energia_Capacitiva_TCP_{ta}', f'Alarma Energia Capacitiva TCP {ta}',          11),
            (f'{p}_Alarma_Sobrecompensacion_TCP_{ta}',  f'Alarma Sobrecompensacion TCP {ta}',           12),
            (f'{p}_Alarma_Subcompensacion_TCP_{ta}',    f'Alarma Subcompensacion TCP {ta}',             13),
        ]:
            filas.append(make_row(name, text, f'{p}_Alarmas', f'{ta}.Alarm2', bit))

    return filas
```

```python
# ------------------------------------------------------------------------------
# FUNCIÓN: insertar filas en el DataFrame después de una fila de referencia
# ------------------------------------------------------------------------------
def insertar_despues_de(df, nombre_ref, nuevas_filas):
    idx = df.index[df['Name'] == nombre_ref].tolist()
    if not idx:
        print(f"  ⚠️  Referencia no encontrada: '{nombre_ref}' — bloque omitido")
        return df
    pos = idx[0] + 1
    df_nuevas = pd.DataFrame(nuevas_filas)
    return pd.concat(
        [df.iloc[:pos], df_nuevas, df.iloc[pos:]],
        ignore_index=True
    )

# ------------------------------------------------------------------------------
# GENERACIÓN PRINCIPAL — recorre todo el diccionario DATA
# ------------------------------------------------------------------------------
todas_las_filas = []

for ps_key, ps_data in DATA.items():
    for ta_key, ta_data in ps_data.items():
        if ta_key == "NamePis":
            continue
        is_tcp    = ta_data["IsTCP"].lower() == "true"
        cant_circ = int(ta_data["CantCirc"])
        todas_las_filas.extend(
            generar_filas_tablero(NOMB_PREFIJO, ta_key, is_tcp, cant_circ)
        )

df_result = pd.DataFrame(todas_las_filas)

# ------------------------------------------------------------------------------
# VARIABLES ESPECIALES DE PISCINA — para TODAS las PS del diccionario
#
# Se insertan después de Comm_Ant del PRIMER tablero de cada piscina.
# Ejemplo: para PS27 → después de CRI-AGROB_Comm_Ant_PS27_TA001
# ------------------------------------------------------------------------------
print("\n📌 Insertando variables especiales de Piscina...")

for ps_key, ps_data in DATA.items():
    # Obtener lista de tableros (excluir la clave "NamePis")
    tableros = [k for k in ps_data.keys() if k != "NamePis"]

    # Punto de inserción: después de Comm_Ant del primer tablero de esta PS
    primer_ta  = tableros[0]
    nombre_ref = f'{NOMB_PREFIJO}_Comm_Ant_{primer_ta}'

    filas_piscina = generar_filas_piscina(NOMB_PREFIJO, ps_key, tableros)
    filas_antes   = len(df_result)
    df_result     = insertar_despues_de(df_result, nombre_ref, filas_piscina)
    filas_despues = len(df_result)

    if filas_despues > filas_antes:
        print(f"  ✅ {ps_key}: {filas_despues - filas_antes} variables insertadas "
              f"(3 control + {len(tableros)} Desactiva_TA) después de '{nombre_ref}'")

# ------------------------------------------------------------------------------
# RENUMERACIÓN DE IDs
# ------------------------------------------------------------------------------
df_result['ID'] = range(START_ALARM_ID, START_ALARM_ID + len(df_result))
```

```python
# ------------------------------------------------------------------------------
# VERIFICACIÓN FINAL
# ------------------------------------------------------------------------------
total_ps = len([k for k in DATA.keys()])
print(f"\n{'='*65}")
print(f"📊  Total alarmas generadas : {len(df_result)}")
print(f"🏊  Piscinas procesadas     : {total_ps}")
print(f"📋  Variables por piscina   : 3 control + N Desactiva_TA (1 por TA)")

# Mostrar resumen de variables de piscina generadas
piscina_vars = df_result[
    df_result['Name'].str.contains('Remoto_Piscina|Selector_Horario|Desactiva_TA_')
]
print(f"\n🔍  Total variables especiales Piscina: {len(piscina_vars)}")
print(piscina_vars[['ID', 'Name', 'Trigger tag']].to_string(index=False))
print(f"{'='*65}\n")

# ------------------------------------------------------------------------------
# GUARDAR EN EXCEL
# ------------------------------------------------------------------------------
df_result.to_excel(OUTPUT_XLSX, sheet_name=SHEET_NAME, index=False)
print(f"💾  Archivo guardado en:\n    {OUTPUT_XLSX}")
```
