```python
# VISUALIZAR LAS COLUMNAS Y HOJAS DEL EXCEL
xls = pd.ExcelFile(PLANTILLA_XLSX)
print(xls.sheet_names)
tpl = pd.read_excel(PLANTILLA_XLSX, sheet_name=xls.sheet_names[0], header=0)
print(tpl.columns.tolist())
```

```python
# =========================
# DEFINICIONES DE ALARMAS
# =========================
ANALOG_DEFS = [
    dict(
        key="Comunic_ANT",
        name_tpl="{F}_Comunic_ANT_{PS}_{TA}",
        text_tpl="Alarma Perdida Comun ANT {PS}_{TA}",
        class_tpl="{F}_Umbral_Antenas",
        trig_tag_tpl="Ant_Pot_{PS}_{TA}",
        model_contains="Comunic_ANT",
    ),
    dict(
        key="Selector_Local_Automatico",
        name_tpl="{F}_Selector_Local_Automatico_{PS}_{TA}",
        text_tpl="Selector Operacion Modo Automático {PS}_{TA}",
        class_tpl="{F}_Operacion",
        trig_tag_tpl="Op_Local_{PS}_{TA}",
        model_contains="Selector_Local_Automatico",
    ),
    dict(
        key="Selector_Local_Manual",
        name_tpl="{F}_Selector_Local_Manual_{PS}_{TA}",
        text_tpl="Selector Operacion Modo Manual {PS}_{TA}",
        class_tpl="{F}_Operacion",
        trig_tag_tpl="Op_Local_{PS}_{TA}",
        model_contains="Selector_Local_Manual",
    ),
    dict(
        key="Selector_Local_Off",
        name_tpl="{F}_Selector_Local_Off_{PS}_{TA}",
        text_tpl="Selector Operacion Modo Off {PS}_{TA}",
        class_tpl="{F}_Operacion",
        trig_tag_tpl="Op_Local_{PS}_{TA}",
        model_contains="Selector_Local_Off",
    ),
]
```

```python
import pandas as pd

# =========================
# HELPERS PARA TU DATA
# =========================
def iter_tas(ps: str, ps_obj: dict):
    """
    Devuelve tuplas (ta_full, ta_short) para cada tablero.
    - ta_full  : "PS27_TA001"
    - ta_short : "TA001"
    Salta "NamePis" u otros metadatos.
    """
    for k, v in ps_obj.items():
        if k == "NamePis":
            continue
        if not isinstance(v, dict):
            continue

        ta_full = k
        #Normaliza: "PS27_TA001" -> "TA001"
        prefix = ps + "_"
        ta_short = ta_full[len(prefix):] if ta_full.startswith(prefix) else ta_full

        yield ta_full, ta_short
# =========================================================
# LECTURA PLANTILLA (ANALOG)
# =========================================================
def load_template():
    xls = pd.ExcelFile(PLANTILLA_XLSX)

    if SHEET_ALARMS not in xls.sheet_names or SHEET_LIMITS not in xls.sheet_names:
        raise ValueError(
            f"La plantilla debe tener hojas '{SHEET_ALARMS}' y '{SHEET_LIMITS}'. "
            f"Hojas encontradas: {xls.sheet_names}"
        )

    df_alarms = pd.read_excel(PLANTILLA_XLSX, sheet_name=SHEET_ALARMS, header=0)
    df_limits = pd.read_excel(PLANTILLA_XLSX, sheet_name=SHEET_LIMITS, header=0)

    df_alarms.columns = df_alarms.columns.astype(str).str.strip()
    df_limits.columns = df_limits.columns.astype(str).str.strip()

    return df_alarms, df_limits


# =========================================================
# MODELOS
# =========================================================
def pick_model_alarm_rows(df_alarms):
    models = {}
    names = df_alarms["Name"].astype(str)

    for d in ANALOG_DEFS:
        key = d["key"]
        m = names.str.contains(d["model_contains"], na=False)
        models[key] = df_alarms[m].iloc[0].copy() if m.any() else df_alarms.iloc[0].copy()

    return models

def pick_model_limits(df_limits):
    model_by_alarm_id = {}
    for _, r in df_limits.iterrows():
        aid = int(r["Alarm ID"])
        model_by_alarm_id[aid] = r.copy()
    return model_by_alarm_id

def make_alarm_row(model_alarm_row, new_alarm_id, name, text, alarm_class, trig_tag, trigger_bit):
    r = model_alarm_row.copy()
    r["ID"] = int(new_alarm_id)
    r["Name"] = name
    r["Alarm text [en-US], Alarm text 1"] = text
    r["Class"] = alarm_class
    r["Trigger tag"] = trig_tag
    # Alarm parameter 1: Tag con Trigger_bit
    r["Alarm parameter 1"] = ""
    # Alarm parameter 2: Tag solo
    r["Alarm parameter 2"] = "" # Changed to empty string
    # Alarm parameter 3: Prefijo nombre
    r["Alarm parameter 3"] = ""          # Changed to empty string
    return r

def make_limit_row(model_limit_row, new_alarm_id):
    r = model_limit_row.copy()
    r["Alarm ID"] = int(new_alarm_id)
    return r

def generate():
    tpl_alarms, tpl_limits = load_template()
    model_alarm_rows = pick_model_alarm_rows(tpl_alarms)
    model_limits_by_id = pick_model_limits(tpl_limits)

    MODEL_ID_BY_KEY = {
        "Comunic_ANT": 1,
        "Selector_Local_Automatico": 2,
        "Selector_Local_Manual": 3,
        "Selector_Local_Off": 4,
    }

    out_alarm_rows = []
    out_limit_rows = []

    # ✅ FORZAR que use el START_ALARM_ID global
    next_id = int(START_ALARM_ID)

    first_id = next_id  # para debug

    for ps, ps_obj in DATA.items():
        for ta_full, ta_short in iter_tas(ps, ps_obj):
            for d in ANALOG_DEFS:
                key = d["key"]

                name = d["name_tpl"].format(F=NOMB_PREFIJO, PS=ps, TA=ta_short)
                text = d["text_tpl"].format(F=NOMB_PREFIJO, PS=ps, TA=ta_short)
                alarm_class = d["class_tpl"].format(F=NOMB_PREFIJO)
                trig_tag = d["trig_tag_tpl"].format(PS=ps, TA=ta_short)

                out_alarm_rows.append(
                    make_alarm_row(
                        model_alarm_rows[key],
                        next_id,
                        name,
                        text,
                        alarm_class,
                        trig_tag,
                        ""   # análogas no usan Trigger bit
                    )
                )

                model_id = MODEL_ID_BY_KEY[key]
                out_limit_rows.append(
                    make_limit_row(model_limits_by_id[model_id], next_id)
                )

                next_id += 1

    last_id = next_id - 1

    out_alarms_df = pd.DataFrame(out_alarm_rows, columns=tpl_alarms.columns)
    out_limits_df = pd.DataFrame(out_limit_rows, columns=tpl_limits.columns)

    with pd.ExcelWriter(OUTPUT_XLSX, engine="openpyxl") as w:
        out_alarms_df.to_excel(w, sheet_name=SHEET_ALARMS, index=False)
        out_limits_df.to_excel(w, sheet_name=SHEET_LIMITS, index=False)

    # ✅ CONFIRMACIÓN EN CONSOLA
    print("START_ALARM_ID esperado:", START_ALARM_ID)
    print("ID generado (primero):", first_id)
    print("ID generado (ultimo):", last_id)

    return OUTPUT_XLSX, len(out_alarms_df), len(out_limits_df)


if __name__ == "__main__":
    out, nA, nL = generate()
    print("OK ->", out)
    print("Filas AnalogAlarms:", nA)
    print("Filas Limits:", nL)
```
